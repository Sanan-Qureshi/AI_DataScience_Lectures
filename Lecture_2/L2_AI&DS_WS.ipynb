There are two methods that can be used to design datasets. <br> One, you can manually enter a dataset into Excel sheet to make a csv file. <br> Another, you can create a dataset manually into Python interface.

```python
import pandas as pd
```

```python
Subject = {
    'Favourite Subject' : ['Maths', 'Science', 'English', 'History']
}
df_1 = pd.DataFrame(Subject)
```

```python
df_1
```

```python
from sklearn.preprocessing import LabelEncoder


# Encode Favourite_Subject
le_subject = LabelEncoder()
df_1['Subject_Code'] = le_subject.fit_transform(df_1['Favourite Subject'])
```

```python
df_1
```

But there is another method as well. <br> What if we want to assign number manually?

```python
Subject_1 = {
    'Favourite Subject' : ['Maths', 'Science', 'English', 'History']
}
df_2 = pd.DataFrame(Subject_1)
```

```python
subject_mapping = {
    'Maths': 2,
    'Science': 3,
    'English': 0,
    'History': 1
}
df_2['Subject_Code'] = df_2['Favourite Subject'].map(subject_mapping)
```

```python
df_2
```

```python
Subject_2 = {
    'Favourite Subject' : ['Maths', 'Science', 'English', 'History']
}
df_3 = pd.DataFrame(Subject_2)
```

```python
df_encoded = pd.get_dummies(df_3, columns=['Favourite Subject'])
```

```python
df_encoded
```

```python
data = {
    'Name':['Ayaan', 'Meera', 'Zara', 'Kabir', 'Priya'],
    'Maths': [88, 67, 45, 92, 70]

}
df_4 = pd.DataFrame(data)
```

```python
df_4
```

```python
# Segmentation
def segment_by_score(score):
    if score < 50:
        return 'Low'
    elif score < 80:
        return 'Medium'
    else:
        return 'High'

df_4['Score_Level'] = df_4['Maths'].apply(segment_by_score)
```
