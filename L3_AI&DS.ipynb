```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Class_Marks_Story.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ðŸŽ¨ Plot the marks to show spread
plt.figure(figsize=(12, 6))
sns.histplot(df['Maths_Marks'], bins=10, kde=True, color='skyblue')
plt.axvline(df['Maths_Marks'].mean(), color='red', linestyle='--', label='Mean')
plt.axvline(df['Maths_Marks'].median(), color='green', linestyle='-.', label='Median')
plt.title("Distribution of Maths Marks")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ðŸŽ¨ Show only distribution (histogram)
plt.figure(figsize=(10, 5))
sns.histplot(df['Maths_Marks'], bins=10, color='skyblue')
plt.title(" Marks Distribution - Just the Histogram")
plt.xlabel("Maths Marks")
plt.ylabel("Number of Students")
plt.grid(True)
plt.show()
```

```python
# ðŸ“ Show histogram + Mean
plt.figure(figsize=(10, 5))
sns.histplot(df['Maths_Marks'], bins=10, color='skyblue')
plt.axvline(df['Maths_Marks'].mean(), color='red', linestyle='--', label='Mean')
plt.title("ðŸ“Š Histogram with Mean")
plt.xlabel("Maths Marks")
plt.ylabel("Number of Students")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ðŸ“ Show histogram + Mean + Median
plt.figure(figsize=(10, 5))
sns.histplot(df['Maths_Marks'], bins=10, color='skyblue')
plt.axvline(df['Maths_Marks'].mean(), color='red', linestyle='--', label='Mean')
plt.axvline(df['Maths_Marks'].median(), color='green', linestyle='-.', label='Median')
plt.title("ðŸ“Š Histogram with Mean and Median")
plt.xlabel("Maths Marks")
plt.ylabel("Number of Students")
plt.legend()
plt.grid(True)
plt.show()
```

Do most students have marks around the average (mean)?

```python
mean = df['Maths_Marks'].mean()
median = df['Maths_Marks'].median()
mode = df['Maths_Marks'].mode()[0]

print(f" Mean (Average): {mean}")
print(f" Median (Middle Value): {median}")
print(f" Mode (Most Frequent Score): {mode}")
```

```python
variance = df['Maths_Marks'].var()
std_dev = df['Maths_Marks'].std()

print(f"ðŸ“ˆ Variance: {variance:.2f}")
print(f"ðŸ“‰ Standard Deviation: {std_dev:.2f}")
```

Variance tells us how spread out the marks are from the mean (average).

If all students scored around 50, the variance would be small.

But if some scored 15 and others scored 100, the variance would be big.

A big number = the scores are very spread out

Standard Deviation tells us: How far, roughly, most marks are from the mean.

So if the mean score is 61, and the standard deviation is 27,
then most scores are somewhere between:

```python
# Count pass/fail
pass_counts = df['Passed'].value_counts(normalize=True) #

# Convert to percentage
pass_percent = pass_counts * 100

print("âœ… Probability of passing (in %):")
print(pass_percent)
```

```python
# Mark outliers: marks < 25 or > 90
df['Outlier'] = df['Maths_Marks'].apply(lambda x: "Yes" if x < 25 or x > 90 else "No")

plt.figure(figsize=(12, 6))
sns.scatterplot(x='Student_ID', y='Maths_Marks', hue='Outlier', data=df, palette={'Yes':'red', 'No':'green'}, s=100)
plt.title("Outliers in Class Performance")
plt.xlabel("Student ID")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```
