```python
import pandas as pd
```

```python
#Let us upload the dataset
df = pd.read_csv("/content/student_data.csv")
```

```python
#and visualise the data
df
```

```python
#Let's start by fixing the 'Age' column.
df['Age'] = df['Age'].fillna(df['Age'].mode()[0])
```

```python
df
```

```python
#Then, we need to fill the missing value by taking mean
df['Maths'] = df['Maths'].fillna(df['Maths'].mean())
```

```python
df['Science'] = df['Science'].fillna(df['Science'].mean())
```

```python
df
```

```python
#we cannot add manually the subject name so we will add 'Unknown' for the missing detail
df['Favourite_Subject'] = df['Favourite_Subject'].fillna('Unknown')
```

```python
#Now we need to normalise our data
df['Maths_Normalised'] = (df['Maths'] - df['Maths'].min()) / (df['Maths'].max() - df['Maths'].min())
df['Science_Normalised'] = (df['Science'] - df['Science'].min()) / (df['Science'].max() - df['Science'].min())
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode Gender
le_gender = LabelEncoder()
df['Gender_Code'] = le_gender.fit_transform(df['Gender'])

# Encode Favourite_Subject
le_subject = LabelEncoder()
df['Subject_Code'] = le_subject.fit_transform(df['Favourite_Subject'])
```

```python
df
```

```python
# Create average marks column
df['Average'] = (df['Maths'] + df['Science']) / 2

# Segment students
def performance_group(score):
    if score >= 0.85:
        return 'High'
    elif score >= 0.7:
        return 'Medium'
    else:
        return 'Low'

df['Performance'] = df['Average'].apply(performance_group)
```
